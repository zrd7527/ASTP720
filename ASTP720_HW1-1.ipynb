```python
'''
ASTP-720 Homework 1
Zach Diermyer
Problem 1
'''

def bisection(func, epsilon, iterate):
    '''
    Finds root of a function using bisection method
    Inputs:
        func - Function to find root of
        epsilon - Threshold value of function to break out of loop and approximate root
        iterate - Boolean, true to print the number of iterations before root is found
    Returns:
        c - Determined value of root
    '''
    a = 11
    b = 11
    for i in range(-10,10):         #Searches for root inside of this range
        if func(i) > 0:             #Assigns positive side of root to a and negative side to b
            a = i
        elif func(i) < 0:
            b = i
        else:                       #If root is an integer, no bisection needed
            print('Root Found: ' + str(i))
            if iterate == True:
                print('Number of Iterations: 0')
            return(i)
        if a < 11 and b < 11:
            break
    c = (a+b)/2
    count = 1
    if abs(func(c)) < epsilon:
        print('Root Found: ' + str(c))
        if iterate == True:
            print('Number of Iterations: 1')
    else:
        while abs(func(c)) > epsilon:
            if func(a)*func(c) < 0:
                b = c
            else:
                a = c
            c = (a+b)/2
            count += 1
        print('Root Found: ' + str(c))
        if iterate == True:
            print('Number of Iterations: ' + str(count))
    return(c)

def newton(func, funcprime, epsilon, iterate):
    '''
    Finds root of a function using Newton's method
    Inputs:
        func - Function to find root of
        funcprime - Derivative of func, input function
        epsilon - Threshold value of function to break out of loop and approximate root
        iterate - Boolean, true to print the number of iterations before root is found
    Returns:
        xn - Determined value of root
    '''
    x0 = -10          #Start search at initial guess value
    xn = x0 - (func(x0)/funcprime(x0))
    if abs(func(xn)) < epsilon:
        print('Root Found: ' + str(xn))
        if iterate == True:
            print('Number of Iterations: 1')
    else:
        count = 1
        while abs(func(xn)) > epsilon:
            xn1 = xn - (func(xn)/funcprime(xn))
            xn = xn1
            count += 1
        print('Root Found: ' + str(xn))
        if iterate == True:
            print('Number of Iterations: ' + str(count))
    return(xn)

def secant(func, epsilon, iterate):
    '''
    Finds root of a function using the secant method
    Inputs:
        func - Function to find root of
        epsilon - Threshold value of function to break out of loop and approximate root
        iterate - Boolean, true to print the number of iterations before root is found
    Returns:
        xn - Determined value of root
    '''
    x0 = -10                   #Start search at initial guess values
    x1 = 11
    xn = x1 - func(x1)*((x1-x0)/(func(x1)-func(x0)))
    if abs(func(xn)) < epsilon:
        print('Root Found: ' + str(xn))
        if iterate == True:
            print('Number of Iterations: 1')
    else:
        count = 1
        xn0 = x1
        while abs(func(xn)) > epsilon:
            xn1 = xn - func(xn)*((xn-xn0)/(func(xn)-func(xn0)))
            xn0 = xn
            xn = xn1
            count += 1
        print('Root Found: ' + str(xn))
        if iterate == True:
            print('Number of Iterations: ' + str(count))
    return(xn)

def sqrt45(x):
    #Returns the square root of 45 as a function
    return(((x**2)-45))

def sqrtprime(x):
    #Returns the derivative of the square root function
    return(2*x)

def main():
    print('Wlecome to Root Finder')
    q1 = input('Would you like to enter a threshold value? ')
    if q1 == 'yes':
        ep = float(input('What threshold value is desired? '))
    else:
        ep = 0.001
    print('Threshold value: ' + str(ep))
    q2 = input('Would you like the number of iterations printed? ')
    if q2 == 'yes':
        it = True
    else:
        it = False
    q3 = input('Choose a method: bisection, newton, secant. ')
    if q3 == 'bisection':
        print('Bisection method selected')
        an = bisection(sqrt45, ep, it)
    elif q3 == 'newton':
        print('Newtonian method selected')
        an = newton(sqrt45, sqrtprime, ep, it)
    else:
        print('Secant method selected')
        an = secant(sqrt45, ep, it)
main()
```

```python
'''
ASTP-720 Homework 1
Zach Diermyer
Problem 3
'''
import matplotlib.pyplot as plt
import numpy as np

def secant(func, epsilon, xprime):
    '''
    Finds root of a function using the secant method
    Inputs:
        func - Function to find root of
        epsilon - Threshold value of function to break out of loop and approximate root
    Returns:
        xn - Determined value of root
        count - Number of iterations required to find root
    '''
    x0 = -1
    x1 = 2
    xn = x1 - func(x1, xprime)*((x1-x0)/(func(x1, xprime)-func(x0, xprime)))
    count = 1
    if abs(func(xn, xprime)) < epsilon:
        pass
    else:
        xn0 = x1
        while abs(func(xn, xprime)) > epsilon:
            xn1 = xn - func(xn, xprime)*((xn-xn0)/(func(xn, xprime)-func(xn0, xprime)))
            xn0 = xn
            xn = xn1
            count += 1
    return(xn, count)

def gaussianlens(x, xprime):
    #Returns Gaussian plasma lens column density equation with given values
    wl = 21
    a = 1.496e13
    D = 3.086e21
    N0 = 3.086e16
    re = 2.817e-13
    constant = ((wl**2)*re*N0*D)/(np.pi*(a**2))
    return(xprime - x*(1 + constant*np.exp(-((x/a)**2))))

def main():
    observer = []
    lensplane = []
    source = [10, 2]
    rays = []
    for i in range(0, 20):
        observer.append([i, 0])
        s0 = secant(gaussianlens, epsilon = 0.0001, xprime = i)
        lensplane.append([s0[0], 1])
        ray = [observer[i], lensplane[i], source]
        rays.append(ray)
        plt.plot(ray, [0,1,2])
    plt.title('Ray Paths from Source to Observer')
    plt.xlabel('Observer Position (AU*10)')
    plt.ylabel('Distance (kpc)')
main()
```
